```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://imarticus.org/'
```

```python
result=requests.get(url)
page_content=result.content
sl_soup=BeautifulSoup(page_content,'html.parser')
```

```python
result.close()
```

```python
for href in sl_soup.findAll('a',href=True):
    print(href['href'])
```
